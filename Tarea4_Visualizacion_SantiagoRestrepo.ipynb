### <font color='46B8A9'> **Gráfico 1** </font>

```python
# Gráfico 1 - Gráfico de barras agrupadas con 2 bins por barra (1 para goles y 1 para asistencias) para los jugadores que más goles y asistencias aportaron

import pandas as pd

# Filtrar solo por nombre, goles y asistencias
columnas_deseadas = ['player', 'goals', 'assists']
df_asgo=players[columnas_deseadas].copy()

# Crear columna suma para identificar a los jugadores que más aportaron
df_asgo['suma']=df_asgo['goals']+df_asgo['assists']

#Cambiar el nombre de las columnas a español
df_asgo = df_asgo.rename(columns={'goals': 'goles','assists':'asistencias'})

# Ordenar el dataframe y extraer los primeros 5 jugadores
data_ordenada = df_asgo.sort_values(by='suma', ascending=False).head(5)

# Gráfica de barras agrupadas utilizando plotly

import plotly.express as px

# Creación de la figura
fig = px.bar(data_ordenada, x='player', y=['goles', 'asistencias'], barmode='group',
             labels={'value': 'Cantidad', 'variable': 'Estadística'},
             title='Cantidad de goles y asistencias por jugador',
             color_discrete_sequence=['#2F4F4F','#FFD700'],
             hover_data={'player':False})

# Ejes
fig.update_xaxes(title_text='Jugador')
fig.update_yaxes(gridcolor='rgba(0, 0, 0, 0.2)',gridwidth=2, griddash='dot')

# Personalizar
fig.update_layout(title_x=0.5,
                  title_font_color='black',
                  height=750,
                  font_family="Verdana",
                  hovermode='x',
                  legend=dict(x=0.78, y=0.99, traceorder='normal', orientation='h'),
                  font=dict(size=14),
                  bargap=0.3,
                  bargroupgap=0.03,
                  plot_bgcolor='rgba(0, 0, 0, 0)')

# Mostrar el gráfico
fig.show()
```

### <font color='46B8A9'> **Gráfico 2** </font>

```python
# Gráfico 2 - Gráfico de burbujas (dispersión) para identificar el aporte de los clubes a las selecciones

import plotly.express as px
import pandas as pd
import plotly.graph_objects as go

# Filtrar solo por nombre, club y seleccion
columnas = ['player', 'team', 'club']
df_pacl=players[columnas].copy()

# Seleccionar solo los clubes que mas se repiten
clubes_mas_comunes = df_pacl['club'].value_counts().head(6).index.tolist()
df_pacl = df_pacl[df_pacl['club'].isin(clubes_mas_comunes)]

df_pacl = df_pacl.groupby(['club', 'team']).size().reset_index(name='num_jugadores')

# Crear un gráfico de dispersión con Plotly Express (burbujas)
fig1 = px.scatter(df_pacl, x='club', y='team', size='num_jugadores',color='num_jugadores',
                 title='Número de jugadores que cada club aporta a cada selección',
                 labels={'club': 'Club', 'team': 'Selección', 'num_jugadores': 'Número de Jugadores'},
                 )

# Personalizar figura
fig1.update_layout(
    title_x=0.5,
    title_font_color='black',
    height=750,
    font_family="Verdana",
    plot_bgcolor='rgba(0, 0, 0, 0)',  # Establecer el fondo del gráfico como transparente
    xaxis=dict(
        gridcolor='rgba(0, 0, 0, 0.2)',
        gridwidth=1,
        griddash='dot'
    ),
    yaxis=dict(
        gridcolor='rgba(0, 0, 0, 0.2)',
        gridwidth=1,
        griddash='dot'
    )
)

# Calcular el total de jugadores por club
total_jugadores_por_club = df_pacl.groupby('club')['num_jugadores'].sum()

# Agregar etiquetas con el total de jugadores por club
for club, total_jugadores in total_jugadores_por_club.items():
    fig1.add_trace(
        go.Scatter(
            x=[club],
            y=['Total Jugadores'],
            text=[f'Total: {total_jugadores}'],
            mode='text',
            showlegend=False
        )
    )

# Ajustar el borde del marcador
fig1.update_traces(marker=dict(line=dict(color='black', width=1)))

# Mostrar el gráfico
fig1.show()
```
